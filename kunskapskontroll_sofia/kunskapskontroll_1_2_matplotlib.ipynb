### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Plotting can be done using either the pyplot interface or the object-oriented interface in Matplotlib. The recommended approach is the object-oriented interface, as it provides more control and customization over the plots. // Sofia

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

In Matplotlib a figure is the top-level container that holds all the plot elements. An axe is a subplot or panel in the figure. An axis is the number line used to represent the data. An artist is the object used to create the plots. // Sofia

### When plotting in Matplotlib, what is the expected input data type?

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.inspect function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 100)    # 100 evenly spaced numbers from -4 to 4 inclusive 
y = x**2                   # define the function to plot

# Pyplot Approach
plt.plot(x, y)   # plot the function and add title and axis labels
plt.title("Plot of y = x^2 (Pyplot)") 
plt.xlabel("Label for x-axis")
plt.ylabel("Label for y-axis")
plt.show() 

# Object-oriented approach
fig, ax = plt.subplots()  # create figure and axis objects, add title and axis labels
ax.plot(x, y)
ax.set_title("Plot of y = x^2 (Object-oriented)")
ax.set_xlabel("Label for x-axis")
ax.set_ylabel("Label for y-axis")
plt.show()
```

Created the figure using both the pyplot approach and the object-oriented approach. The pyplot approach uses the plt.plot() function, while the object-oriented approach uses the fig, ax = plt.subplots() method. // Sofia

```python
## Using the provided date from the assignment above and adjusted it accordingly

import matplotlib.pyplot as plt
import numpy as np

# Scatter plot data
np.random.seed(15)  # set seed for reproducibility
random_data_x = np.random.randn(1000) # generate random data
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100) # 100 evenly spaced numbers from -2 to 2 inclusive
y = x**2

# Bar plot data
fruit_data = {"grapes": 22, "apple": 8, "orange": 15, "lemon": 20, "lime": 25} 
names = list(fruit_data.keys())  # define names and keys
values = list(fruit_data.values()) # define values

# Create a figure with 2 subplots, one scatter plot and one bar plot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # create a figure and axis objects

# Scatter plot
axs[0].scatter(random_data_x, random_data_y)  # scatter plot
axs[0].set_title("Scatter plot")  # add title and axis labels to the figure
axs[0].set_xlabel("Label for x-axis")
axs[0].set_ylabel("Label for y-axis")

# Bar plot
axs[1].bar(names, values)  # bar plot
axs[1].set_title("Fruit count")

# Add title to the figure and display it
fig.suptitle("Scatter plot and bar plot")
plt.show()


```

Created a figure with two subplots with the provided data. The first subplot is a scatter plot, and the second subplot is a bar plot. The scatter plot is created using random data, and the bar plot is created using the provided fruit data. The figure is displayed using the plt.show() function. // Sofia
